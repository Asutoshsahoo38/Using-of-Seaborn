```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Toyota.csv",index_col=0,na_values=['??','????'])
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.dropna(axis=0, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
sns.set(style='darkgrid') ##for background styling
sns.scatterplot(x=data['Age'],y=data['Price'])
```

```python
sns.regplot(x=data['Age'],y=data['Price']) ## it shows that by increasing the age the price of the car decreases
## Regression Plot
```

```python
sns.regplot(x=data['Age'],y=data['Price'],fit_reg=False) 
```

```python
sns.regplot(x=data['Age'],y=data['Price'],marker="*",fit_reg=False) 
```

```python
sns.lmplot(x='Age',y='Price',data=data,hue='Automatic',fit_reg=False,legend=True) ##hue is used for showing the difference
##by a catagory
```

```python
## Histogram plot 
sns.displot(data['Age'],kde=False,bins=5)
```

```python
sns.displot(data['Age'],kde=True,bins=5)
```

```python
## Bar plot is used for catagorical features such as FuelType 
sns.countplot(x ="FuelType",data = data);
```

```python
sns.countplot(x ="FuelType",data = data,hue = "Automatic");
```

```python
sns.boxplot(x =data['Price']);
```

```python
sns.boxplot(y =data['Price']);
```

```python
## box plot is use to compare b|w catagorical and numerical value
sns.boxplot(y =data['Price'],x=data['FuelType']);
```

```python
sns.boxplot(y =data['Price'],x=data['FuelType'],hue = data["Automatic"]);
```

```python
f,(ax_box,ax_hist)= plt.subplots(2,gridspec_kw={"height_ratios":(.15,.85)});
```
